```python
# Exercício 5-5 e 5-6: média dos pares e média dos ímpares
print('Digite 10 númros quaisquer:')
print('Ao final direi a média dos pares e dos ímpares.')
número_par=0
número_ímpar=0
soma_pares=0
soma_ímpares=0
for i in range (1,11):
    número=int(input('[{0}]:'.format(i)))
    if número%2==0:
        número_par+=1
        soma_pares+=número
    else:
        número_ímpar+=1
        soma_ímpares+=número
if número_par>0:
    print('A média dos pares é:', soma_pares/número_par)
else:
    print('Não há números pares.')
if número_ímpar>0:
    print('A média dos ímpares é: ',soma_ímpares/número_ímpar)
else:
    print('Não há números ímpares.')
```

## seção 5-6 - Variável Booleana

```python
x=int(input('x='))
y=int(input('y='))
if x>y:
    print('x maior que y.')
elif x==y:
    print('x é igual a y.')
else:
    print('x é menor que y.')
```

```python
# O número é primo ou não?
número=int(input('Digite um número e te direi se é primo:'))
é_primo=True
for possível_divisor in range(2,número//2+1):
    if número % possível_divisor==0:
        é_primo=False
if é_primo:
    print('É primo.')
else:
    print('Não é primo.')
```

```python
#Formação de triângulos
print('Entre com 3 valores. O maior primeiro.')
a=int(input('a='))
b=int(input('b='))
c=int(input('c='))
if a>=b+c or\
    b>=a+c or\
    c>=a+b:
    print('Não podem formar um triãngulo.')
else:
    print('formam um triângulo.')
```

## seção 5-8 - If-elif-else

```python
x=int(input('Primeiro número:'))
y=int(input('Segundo número:'))
op=input('Operador (+,-,/,*):')
if op=='+':
    print(x+y)
elif op=='-':
    print(x-y)
elif op=='/':
    print(x/y)
elif op=='*':
    print(x*y)
else:
    print('Operador desconhecido.')
```

```python
print('Vote: 11- Abel, 15- Beatriz, 21- Caio, 25- Dirceu, 46-Eustáquio')
votos_Abel=0
votos_Beatriz=0
votos_Caio=0
votos_Dirceu=0
votos_Eustáquio=0
votos_brancos=0
votos_nulos=0
for i in range(1,11):
    voto=input('Vote no seu candidato:')
    if voto=='11':
        votos_Abel+=1
    elif voto=='15':
        votos_Beatriz+=1
    elif voto=='21':
        votos_Caio+=1
    elif voto=='25':
        votos_Dirceu+=1
    elif voto=='46':
        votos_Eustáquio+=1
    elif voto=='0':
        votos_brancos+=1
    else:
        votos_nulos+=1
print('\nResultado da votação\n')
print('votos para o Abel:',votos_Abel)
print('votos para o Beatriz:',votos_Caio)
print('votos para o Caio:',votos_Dirceu)
print('votos para o Dirceu:',votos_Eustáquio)
print('votos brancos:',votos_brancos)
print('votos nulos:',votos_nulos)
```
