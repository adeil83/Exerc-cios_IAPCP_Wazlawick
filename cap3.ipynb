# 3.2 - Comando de atribuição 

O sinal de "=", usado nos comandos de atribuição, não corresponde extamente à igualdade matemática, mas um conceito um pouco diferente.
A diferença que existe entre a notação de igualdade na matemática e o comando de atribuição em python e na maioria das linguagens de programação é que x=y+z representa a operação de colocar em "x" o resultado da soma de "y" e "z". Pense novamente na máquina de Babbage e suas torres de engrenagens; x=y+z é um comando que você dá à máquina, o qual, como vimos antes, faz com que a máquina execute a soma de "y" e "z" e armazena o resutado na variável "x".
Esse comando não é comutativo como a igualdade matemática, ou seja, não se pode escrever y+z=x. Por quê? Basicamente porque o interpretador tentaria, neste caso, colocar o valor da variável "x" na soma de "y" e "z"! pág. 21 e 22.

```python
x=1 #depois deste comando, x vale 1
print(x)
x=x+1 #depois deste comando, x vale 2
print(x)
x=x+10 # depois deste comando, x vale 12
print(x)
```
