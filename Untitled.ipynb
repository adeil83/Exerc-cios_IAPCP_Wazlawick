```python
t = npy.arange(0.0, 6.4, 0.1)
y = 5*npy.sin(t)
z = 3*npy.cos(t)
gesamt = plt.figure() #1
diag1 = gesamt.add_subplot(2,1,1) #2
diag1.plot(t,y,color="red",linewidth=3) #3
diag1.set_title("Sinus-Funktion") #4
diag2 = gesamt.add_subplot(2,1,2) #5
diag2.plot(t,z,color="blue",linewidth=2) #6
diag2.set_title("Kosinus-Funktion") #7
gesamt.show() 
```

```python
t = npy.arange(0.0, 6.4, 0.1)
y = 5*npy.sin(t)
z = 3*npy.cos(t)
plt.plot(t, y, "r+-", linewidth=2.0, label="sin") #1
plt.plot(t, z, "bo-", linewidth=2.0, label="cos") #2
plt.xlabel('Zeit in Sekunden')
plt.ylabel('y(t), z(t)')
plt.title('Die Funktionen y(t)=5*sin(t) und z(t)=3*cos(t)')
plt.legend(("sin","cos")) #3
plt.show()
```

```python
t = npy.arange(0.0, 2.1, 0.1)
y1 = npy.exp(-t) #1
y2 = npy.exp(-2*t) #2
y3 = npy.exp(-4*t) #3
plt.plot(t, y1, "r+-", linewidth=2.0, label="k=1") #4
plt.plot(t, y2, "bo-", linewidth=2.0, label="k=2") #5
plt.plot(t, y3, "g^-", linewidth=2.0, label="k=3") #6
plt.xlabel('Zeit in Sekunden')
plt.ylabel('y(t)')
plt.title('Die Funktion y(t)=exp(-k*t) fuer k=1,2,4')
plt.legend(("k=1","k=2","k=4"))
plt.show()
```

```python
from scipy.interpolate import interp1d #1
import numpy as np #2
import matplotlib.pyplot as plt #3
x = np.linspace(1, 6, 6) #4
y = np.array([1, 3, 1, 5, 4, 4]) #5
xn = np.linspace(1,6,60) #6
f = interp1d(x, y,kind='linear') #7
g = interp1d(x,y,kind='cubic') #8
h = interp1d(x,y,kind=5) #9
plt.plot(x,y,'ro',linewidth=3.0) #10
plt.plot(xn,f(xn),'b-',linewidth=3) #11
plt.plot(xn,g(xn),'g--',linewidth=3) #12
plt.plot(xn,h(xn),'co-',linewidth=3) #13
plt.xlabel("x-Werte") #14
plt.ylabel("y-Werte") #15
plt.title("Interpolation") #16
plt.legend(["Vorgabe","Linear",
 "Kubisch","Polynom Grad 5"],
 loc="best") #17
plt.show() 
```

```python
from scipy.integrate import odeint #1
import numpy as np
import matplotlib.pyplot as plt
def dy_dx(y,x): #2
    return y + x
x = np.linspace(0,6,100) #3
y0 = 0.0 #4
ys = odeint(dy_dx, y0, x) #5
y = np.array(ys).flatten() #6
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x, y,"g-",linewidth=3)
plt.show()
def exakt(x): #7
    return np.exp(x) - x - 1.0
y2 = exakt(x) #8
fehler = y - y2 #9
plt.xlabel("x")
plt.ylabel("Fehler")
plt.grid()
plt.plot(x, fehler,"b-",linewidth=3)
plt.show()
```
