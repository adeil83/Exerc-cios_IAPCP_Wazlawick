```python
# Exercício 4-10
a=int(input('Digite um valor para a:'))
b=int(input('Digite um valor para b:'))
n=int(input('Digite um valor para n:'))
somatório=0
for i in range(1,n+1):
    somatório+=a-b*i+i**2
print(somatório)
```

# Seção 4.9 - Fatorial

```python
n=int(input('Diga um número e direi o fatorial:'))
fatorial=1
for i in range(1,n+1):
    fatorial*=i
    print('{0}!={1}'.format(i,fatorial))
```

# Seção 4-10 - Aninhamento

```python
for j in range (1,11):
    print(4*j)
```

```python
#Tabuada
for i in range (1,11):
    print('\nTabuada de:',i)
    for j in range (1,11):
        print('{0}x{1}={2}'.format(i,j,i*j))
```

```python
# Relógio Digital
for hora in range(24):
    for minuto in range(60):
        for segundo in range(60):
            print('{0}h{1}min{2}s'.format(hora,minuto,segundo))
```

```python
#Calendário
for ano in range(2016,2021):
    for mês in range(1,13):
        for dia in range(1,31):
            print('{0}/{1}/{2}'.format(dia,mês,ano))
            
```
