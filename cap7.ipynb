```python
#Arranjos

def fatorial(n):  # n é o parâmetro
    fat=1
    for i in range(2,n+1):
        fat*=i
    return fat

print('Cálculo de arranjos A(n,p)')
n=int(input('Quantos elementos no total (n)?'))
p=int(input('Quantos elementos em cada arranjo (p)?'))
a=fatorial(n)//fatorial(n-p)
print('Arranjos possíveis:',a)
```

```python
#Exercício 7-3 

#Combinações

def fatorial(n):  # n é o parâmetro
    fat=1
    for i in range(2,n+1):
        fat*=i
    return fat

print('Cálculo de combinações C(n,p)')
n=int(input('Quantos elementos no total (n)?'))
p=int(input('Quantos elementos em cada arranjo (p)?'))
a=fatorial(n)//(fatorial(p)*fatorial(n-p))
print('Combinações possíveis:',a)
```
